Prompt (prompt)

    The input sentence influences the style of the generated text.

    If the prompt is a question, GPT-2 might try to generate an answer.

    If the prompt is a story-like opening, GPT-2 is likely to continue the story.

Model (generator)

    pipeline("text-generation", model="gpt2") loads GPT-2, which has been trained on a large dataset of books, stories, and articles.

    Because GPT-2 learned patterns from books and stories, it often naturally continues a prompt in a narrative style.

max_length=200

    Controls how long the generated text is.

    A longer max_length makes it more likely to generate a story-like continuation rather than a short answer.

num_return_sequences=1

    This specifies how many generated texts we want.

    If you set num_return_sequences=3, it would generate 3 different possible continuations for the same prompt.

#Step 2: Load a Pre-Trained Text Generation Model
You can use GPT-2 (or other models like facebook/bart-large-cnn, bigscience/bloom, etc.).

```python
from transformers import pipeline

# Load the text generation pipeline using GPT-2
generator = pipeline("text-generation", model="gpt2")

# Test with a prompt
prompt = "Once upon a time in a faraway land"
output = generator(prompt, max_length=50, num_return_sequences=1)

# Print the generated text
print(output[0]["generated_text"])
```

# Key Observations from the Generated Output
Continuation of the Story

"She became trapped in it. A certain sort of mystery unfolds, and the girl who entered can tell you much more."

 - GPT-2 understood that a "hidden door" suggests something mysterious.

- It continued the plot with suspense—she is "trapped," and "a mystery unfolds."

Ambiguity and Vagueness

"It may seem a long wait with few answers, but she will learn the full story soon, just like the others."

 - This is a typical GPT-2 output trait—it continues generating text in a plausible but vague manner.

 - It hints at a larger narrative ("the others"), but it doesn’t specify who "the others" are.

Dramatic Build-up

 - "The journey will be long, and the journey will be dangerous. So, do not be afraid of the first day!"

 - GPT-2 introduces tension—it makes the story sound adventurous.

 - It repeats "the journey will be long and dangerous"—showing a bias towards dramatic storytelling.

Strange Transition into a Game-like Format

"It may take three (3) weeks to complete The Darkest Story, but we shall soon be back in action!"

 - This is an unexpected shift in tone—from a mysterious fairy tale to what sounds like a video game or an episodic series.

 - GPT-2 sometimes hallucinates text based on its training data (which likely included game descriptions and interactive stories).

Game Instructions and "Character Story"

"You should be able to play as your character at this point."

 - GPT-2 seems to think this is an interactive fiction or game script.

"We have already put together a list for you in the description below!" → This mimics game patch notes or a story update.

```python
from transformers import pipeline

# Load the text generation pipeline with GPT-2
generator = pipeline("text-generation", model="gpt2")

# Story-like prompt
prompt = "Once upon a time, in a mysterious forest, a young girl discovered a hidden door."

# Generate text with a longer continuation
output = generator(prompt, max_length=150, num_return_sequences=1)

# Print the generated story
print(output[0]["generated_text"])
```

#Streamlit app

```python
%%writefile text_generator_app.py
import streamlit as st
from transformers import pipeline

# Load the text generation pipeline using GPT-2
generator = pipeline("text-generation", model="gpt2")

# Streamlit interface
st.title("Text Generator App")
prompt = st.text_input("Enter a prompt", "Can you give me a famous quote by Nelson Mandela?")

if prompt:
    output = generator(prompt, max_length=200, num_return_sequences=1)
    st.write(output[0]["generated_text"])
```
